```python
df.columns
```

```python
#query fonksiyonunda expression yazrken kolon adın boşluk varsa AltGr + ; ile yazmak gerek.
#Talebin Gerekçesi alanı Bug Fixes ise
df.query('`Talebin Gerekçesi` == "Bug Fix" and `Yıllık Sağlanacak Fayda/Verimlilik (Saat)`!=0')
```

```python
#df.Statü.value_counts()
dfs = df.query("Statü.str.contains('Tamamlandı')", engine='python')
dfs["Yıllık Sağlanacak Fayda/Verimlilik (Saat)"].sum()

#
dfs.groupby(by="Oluşturan Personel")["Yıllık Sağlanacak Fayda/Verimlilik (Saat)"] \
   .agg(["sum","count","max","min","mean"])
```

```python
#df.groupby(by="Oluşturan Departman Adı").count()
df["Oluşturan Departman Adı"].value_counts().to_frame().rename(columns={"count":"adet"}).reset_index().to_excel("./databases/departman.xlsx",index=False)
```

```python
df["Yil_Ay"] = df["Oluşturma Tarihi"].dt.strftime("%Y-%m")
df.columns

#
dfg = df.groupby(by="Yil_Ay").agg(
            {
                "Yıllık Sağlanacak Fayda/Verimlilik (Saat)":sum
             })

dfg
```
